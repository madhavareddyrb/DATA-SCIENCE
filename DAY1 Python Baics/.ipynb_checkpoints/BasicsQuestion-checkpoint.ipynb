**Question 8:** Write a Python program to convert Celsius to Fahrenheit.

```python
# Your code here
choose = input("Enter Your Unit to convert: ")

temprature = 0
if choose == 'C' :
    converttoFa = (temprature * 9/5) + 32
    print(converttoFa)
elif choose == "F" :
    convertToC = (temprature - 32) * 5/9
    print(convertToC)
else:
    print('choose between "c" or "F"')
    
```
